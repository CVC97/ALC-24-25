```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# function to process gromacs files
def process_gromacs_file(gromacs_file, skiprows):
    list_of_rows = []
    for i_row, line in enumerate(open(gromacs_file)):
        if i_row >= skiprows:
            list_of_rows.append(np.array([*line.split()], dtype = float))
    return np.array(list_of_rows)
```

```python
# data processing

# 270 K
pmf_rgyr_270K_data = process_gromacs_file("data/270K/pmf-rg.xvg", 33)
pmf_dist_270K_data = process_gromacs_file("data/270K/pmf-dist.xvg", 33)
rgyr_dist_270K_data = process_gromacs_file("data/270K/xy-rg-dist.xvg", 33)
rgyr_epot_270K_data = process_gromacs_file("data/270K/xy-rg-potential.xvg", 33)


# 300 K - 10 kJ/mol bond angle spring constant
pmf_rgyr_300K_10kJ_data = process_gromacs_file("data/300K_10kJ/pmf-rg.xvg", 33)
pmf_dist_300K_10kJ_data = process_gromacs_file("data/300K_10kJ/pmf-dist.xvg", 33)
rgyr_dist_300K_10kJ_data = process_gromacs_file("data/300K_10kJ/xy-rg-dist.xvg", 33)
rgyr_epot_300K_10kJ_data = process_gromacs_file("data/300K_10kJ/xy-rg-potential.xvg", 33)

angle_distribution_300K_10kJ_data = process_gromacs_file("data/300K_10kJ/angdist.xvg", 33)


# 300 K - 35 kJ/mol bond angle spring constant
pmf_rgyr_300K_35kJ_data = process_gromacs_file("data/300K_35kJ/pmf-rg.xvg", 33)
pmf_dist_300K_35kJ_data = process_gromacs_file("data/300K_35kJ/pmf-dist.xvg", 33)
rgyr_dist_300K_35kJ_data = process_gromacs_file("data/300K_35kJ/xy-rg-dist.xvg", 33)
rgyr_epot_300K_35kJ_data = process_gromacs_file("data/300K_35kJ/xy-rg-potential.xvg", 33)

angle_distribution_300K_35kJ_data = process_gromacs_file("data/300K_35kJ/angdist.xvg", 33)

# 300 K - 45 kJ/mol bond angle spring constant
pmf_rgyr_300K_45kJ_data = process_gromacs_file("data/300K_45kJ/pmf-rg.xvg", 33)
pmf_dist_300K_45kJ_data = process_gromacs_file("data/300K_45kJ/pmf-dist.xvg", 33)
rgyr_dist_300K_45kJ_data = process_gromacs_file("data/300K_45kJ/xy-rg-dist.xvg", 33)
rgyr_epot_300K_45kJ_data = process_gromacs_file("data/300K_45kJ/xy-rg-potential.xvg", 33)

angle_distribution_300K_45kJ_data = process_gromacs_file("data/300K_45kJ/angdist.xvg", 33)


# 330 K
pmf_rgyr_330K_data = process_gromacs_file("data/330K/pmf-rg.xvg", 33)
pmf_dist_330K_data = process_gromacs_file("data/330K/pmf-dist.xvg", 33)
rgyr_dist_330K_data = process_gromacs_file("data/330K/xy-rg-dist.xvg", 33)
rgyr_epot_330K_data = process_gromacs_file("data/330K/xy-rg-potential.xvg", 33)


# 400 K
pmf_rgyr_400K_data = process_gromacs_file("data/400K/pmf-rg.xvg", 33)
pmf_dist_400K_data = process_gromacs_file("data/400K/pmf-dist.xvg", 33)
rgyr_dist_400K_data = process_gromacs_file("data/400K/xy-rg-dist.xvg", 33)
rgyr_epot_400K_data = process_gromacs_file("data/400K/xy-rg-potential.xvg", 33)


# entropic spring
pmf_dist_espring_data = process_gromacs_file("data/espring/pmf-dist.xvg", 33)
```
